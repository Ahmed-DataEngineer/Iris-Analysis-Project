```python
import pandas as pd
```

```python
from matplotlib import pyplot as plt
```

```python
import seaborn as sns
```

```python
data = pd.read_csv("iris.csv")
data
```

```python
sns.set(style="whitegrid")
sns.scatterplot(x="sepal.length",y="petal.length",hue="variety",style="variety",data=data)
plt.show()
```

```python
sns.set(style="whitegrid")
sns.scatterplot(x="sepal.width",y="petal.width",hue="variety",style="variety",data=data)
plt.show()
```
